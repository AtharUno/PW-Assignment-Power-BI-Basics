Question 1: What is Power BI and why is it used in businesses?

Answer:

Power BI is a Business Intelligence and data visualization tool. It allows users to connect to various data sources, clean and transform raw data, and create interactive dashboards, reports, and charts.

Why it is used in businesses:
- Data Visualization: Converts complex numerical datasets into easy-to-understand visual charts and graphs.
- Data Integration: Connects to multiple sources (Excel, SQL databases, Cloud services, Web) simultaneously.
- Real-time Insights: Enables real-time tracking of Business Key Performance Indicators (KPIs).
- Data-driven Decision Making: Helps business leaders and management make informed decisions based on historical and predictive trends.
-Collaboration & Sharing: Allows easy sharing of reports and dashboards across teams securely.

Question 2: Name and explain the three main components of Power BI.

Answer:
The three main components of Power BI are:

1. Power BI Desktop:
A free desktop application used by data analysts to connect to data sources, clean/transform data, model data, and build interactive reports and dashboards.

2. Power BI Service (SaaS / Cloud):
A cloud-based online service where reports created on Power BI Desktop are published, shared, and monitored across organizations.

3. Power BI Mobile:
Mobile apps available for iOS, Android, and Windows devices that allow users to view, interact with, and monitor dashboards securely on the go.

Question 3: Explain the Power BI workflow.

Answer:
The standard Power BI workflow follows a 4-step process:
1. Get Data (Extract): Connect to one or multiple data sources (such as Excel, SQL Server, CSV files, or web pages).

2. Transform Data (Clean & Load): Use Power Query to clean, reshape, combine, and clean the imported data (e.g., removing null values, changing data types).

3. Data Modeling & Visualization: Create relationships between different tables, write DAX (Data Analysis Expressions) measures/columns, and build visual reports using charts, tables, and slicers.

4. Publish & Share: Publish the completed report from Power BI Desktop to the Power BI Service to share dashboards with team members or stakeholders.

Question 4: List any four data cleaning tasks that can be performed in Power Query.

Answer:
Four common data cleaning tasks performed in Power Query are:

1. Removing Duplicates and Null/Blank Values: Eliminating identical rows or empty records to ensure data accuracy.

2. Splitting Columns: Dividing a single column into multiple columns based on a delimiter (e.g., splitting a full name into "First Name" and "Last Name").

3. Changing Data Types: Converting text data types to numbers, dates, or boolean types for accurate calculation and visualization.

4. Replacing Values: Replacing errors, blank fields, or specific text with appropriate default values.

Question 5: Write step-by-step instructions to load the Real_Estate Dataset.

Answer:

Step 1: Open Power BI Desktop.

Step 2: On the Home tab in the ribbon, click on Get Data from csv file.

Step 3: Browse your computer, select the Real_Estate Dataset file, and click Open.

Step 4: In the Navigator window that pops up, check the box next to the table/sheet name containing the real estate data to preview it.

Step 5: Click on Transform Data (to clean it first in Power Query) or Load (to load the raw dataset directly into the Power BI data model).

Question 6: Define Data View, Report View, and Model View. Explain the purpose of each view.

Answer:

Report View:

Definition: The canvas area where you build interactive visual reports.

Purpose: To add charts, graphs, maps, cards, and slicers to present insights visually to end users.

Data View:

Definition: A grid view that displays the actual tabular data loaded into the data model (similar to an Excel spreadsheet).

Purpose: To inspect rows/columns, filter data, create new calculated columns, and write DAX formulas.

Model View:

Definition: A diagram interface that shows all loaded tables as blocks along with their connections.

Purpose: To view, create, edit, and manage relationships (e.g., 1-to-many, many-to-many) between different datasets/tables.

Question 7: Discuss the different data sources that Power BI supports.

Answer:
Power BI supports a wide range of data sources categorized into:

Files: Microsoft Excel (.xlsx), CSV/Text files, XML, JSON, PDF, and local folders.

Databases: SQL Server, Oracle, MySQL, Access.

Power Platform / Cloud Services: Power BI datasets, Power BI datamarts, Dataverse.

Azure / Cloud Databases: Azure SQL Database, Azure Synapse Analytics, Azure Blob Storage.

Online / Web Services: Google Analytics, Salesforce, SharePoint Online lists, OData feeds, and Web URLs.

Question 8: Split Owner Name to create two new columns as First Name and Last Name.
Answer:

Step-by-Step Instructions in Power Query:

In Power BI Desktop, click on Transform Data from the Home tab to open the Power Query Editor.

Select the table containing your dataset from the left panel.

Click on the column header for Owner Name to highlight it.

Go to the Transform or Add Column tab in the top menu bar.

Click on Split Column > By Delimiter.

In the pop-up window:

Select Space as the delimiter.

Choose At the left-most delimiter (or Each occurrence of the delimiter).

Click OK. This will create two new columns named Owner Name.1 and Owner Name.2.

Double-click the header of Owner Name.1 and rename it to First Name.

Double-click the header of Owner Name.2 and rename it to Last Name.

Click Close & Apply on the Home tab to save the changes to your Power BI report.
